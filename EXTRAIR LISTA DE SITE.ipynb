```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.tabeladecalorias.net/alimento/fruta"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

tabela = soup.find("table", {"id": "ID_DA_TABELA"}) # Encontra a tabela pelo ID

dados = []
for linha in tabela.find_all("th"):
    colunas = []
    for coluna in linha.find_all(["th", "td"]):
        colunas.append(coluna.text)
    dados.append(colunas)

df = pd.DataFrame(dados) # Cria um DataFrame com os dados
print(df)
```
